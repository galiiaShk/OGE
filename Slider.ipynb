```python
import numpy as np
import matplotlib.pyplot as plt
def rotate_points(point1, point2, angle):
    # Convert angle to radians
    angle_rad = angle * 2 * np.pi
    
    # Calculate rotation matrix
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
                                [np.sin(angle_rad), np.cos(angle_rad)]])
    
    # Create column matrices for the points
    point1_matrix = np.array([[point1[0]], [point1[1]]])
    point2_matrix = np.array([[point2[0]], [point2[1]]])
    
    # Rotate the points around the origin
    rotated_point1 = np.dot(rotation_matrix, point1_matrix)
    rotated_point2 = np.dot(rotation_matrix, point2_matrix)
    
    return rotated_point1.flatten(), rotated_point2.flatten()

# Define the two points
point1 = np.array([1, 0])
point2 = np.array([1, 0])

# Generate angles from 0 to 1
angles = np.linspace(0, 10, 100)

# Rotate points for each angle
rotated_points = [rotate_points(point1, point2, angle) for angle in angles]

# Unpack rotated points
rotated_point1, rotated_point2 = zip(*rotated_points)
print(rotated_point1[99])
circle1 = plt.Circle(rotated_point1[99],0.2, color='r')
circle2 = plt.Circle(rotated_point1[0], 0.2, color='blue')

fig, ax = plt.subplots() # note we must use plt.subplots, not plt.subplot
# (or if you have an existing figure)
# fig = plt.gcf()
# ax = fig.gca()

ax.add_patch(circle1)
ax.add_patch(circle2)

#plt.scatter(rotated_point1, rotated_point2, c=angles, cmap='viridis')
#plt.colorbar(label='Rotation Angle')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Rotation of Two Points')
plt.show()
```
